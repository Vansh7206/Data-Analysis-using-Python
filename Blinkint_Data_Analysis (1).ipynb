**CHART REQUIREMENTS**


**Sales by Fat**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat, labels= sales_by_fat.index, autopct = '%.1f%%')
plt.title("Sales by fat")

plt.show()
```

**Total Sales by item type**

```python
sales_by_type = df.groupby("Item Type")["Sales"].sum().sort_values(ascending = True)
plt.figure(figsize = (10,6))

plt.pie(sales_by_type, labels = sales_by_type.index, autopct = '%.f%%')
plt.title("Sales by Item Type")

plt.show()
```

**Fat Content by outlet for total sales**

```python
grouped = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]

ax = grouped.plot(kind = 'bar', figsize = (8,5), title = 'Outlet Tier by item Fat content')
plt.xlabel("Outlet location by teir")
plt.ylabel("Total sales")
plt.legend(title = 'Item Fat Content')

plt.tight_layout()
plt.show()
```

**Total Sales by Outlet Establishment**

```python
df.head()
```

```python
sales_by_outletyear = df.groupby('Outlet Establishment Year')["Sales"].sum().sort_index()
plt.figure(figsize=(9,5))
plt.plot(sales_by_outletyear.index,sales_by_outletyear.values, marker = 'o', linestyle = '-')

plt.xlabel('Outlet Establishment year')
plt.ylabel('Total Sales')
plt.title("Outlet Establishment")

for x, y in zip(sales_by_outletyear.index,sales_by_outletyear.values):
    plt.text(x,y,f'{y:,.0f}',ha = 'center', va = 'bottom', fontsize = 8)

plt.tight_layout()

plt.show()
```

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(9,5))

plt.pie(sales_by_size,labels = sales_by_size.index, autopct = '%.1f%%')
plt.title("Sales by outlet size")
plt.show()
```

```python
sales_by_loc = df.groupby('Outlet Location Type')['Sales'].sum()
plt.figure(figsize=(9,5))

plt.pie(sales_by_loc,labels = sales_by_loc.index, autopct = '%.1f%%')
plt.title("Sales by outlet location")
plt.show()
```
